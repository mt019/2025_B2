```python
import re

def extract_sections_by_number(file_path):
    """從 Markdown 中提取以 ## 開頭並編號的區塊，回傳 dict：{編號: 區塊內容}"""
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()

    # 捕捉每個區塊：## 1. 開頭直到下一個 ## 或結尾
    pattern = r'(##\s+(\d+)\.\s+[^\n]+\n(?:.*?))(?=\n##\s+\d+\.|\Z)'
    matches = re.findall(pattern, content, flags=re.DOTALL)

    section_map = {}
    for full_block, num_str in matches:
        num = int(num_str)
        section_map[num] = full_block.strip()

    return section_map


import re

def reorder_and_renumber_sections(section_map, new_order):
    """根據新順序重新排序並重新編號每段，並清除重複分隔線"""
    reordered_sections = []
    for new_index, original_num in enumerate(new_order, 1):
        section = section_map.get(original_num)
        if section:
            # 替換段首的編號為新順序
            section = re.sub(r'^##\s+\d+\.', f'## {new_index}.', section)

            # 清除段落開頭或結尾多餘的分隔線
            section = re.sub(r'^\s*---+\s*', '', section, flags=re.MULTILINE)
            section = re.sub(r'\s*---+\s*$', '', section, flags=re.MULTILINE)
            reordered_sections.append(section.strip())

    # 合併時，僅插入一條分隔線
    return '\n\n---\n\n'.join(reordered_sections)

section_map = extract_sections_by_number(file_path)


```

```python
# 列出 section_map 的 key 與標題內容（僅顯示第一行標題）
for num in sorted(section_map.keys()):
    section = section_map[num]
    # 抓出第一行作為標題（通常是 ## 開頭）
    first_line = section.split('\n')[0]
    print(f"{num}: {first_line}")
```

```python

# 重排與重新編號
reordered_text = reorder_and_renumber_sections(section_map, new_order)

# # 儲存成新檔案
# output_path = "/Users/iw/Documents/NTU/1132/2025_B2/Chapters/00__詞彙表知識點/0504-01__reordered.md"
with open(output_path, 'w', encoding='utf-8') as f:
    f.write(reordered_text)

print(f"✅ 重排並重新編號完成，已儲存至：{output_path}")
```
